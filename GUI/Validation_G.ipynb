# Protocol 1 statistical analysis

```python
# Definisci il target RR
target_rr = 12

# Lista dei dataframe delle varie opzioni
dfs = [df_p1_o1, df_p1_o2, df_p1_o3, df_p1_o4]
option_names = ['Option 1', 'Option 2', 'Option 3', 'Option 4']

# Funzione per eseguire l'analisi statistica e generare il grafico per ogni opzione
def analyze_option(df, option_name):
    df['Error'] = df['RR (breaths/min)'] - target_rr

    # Calcolo delle metriche statistiche
    mse = mean_squared_error(df['RR (breaths/min)'], [target_rr] * len(df))
    mae = mean_absolute_error(df['RR (breaths/min)'], [target_rr] * len(df))
    rmse = np.sqrt(mse)
    std_dev = np.std(df['RR (breaths/min)'])

    # Stampa delle metriche
    print(f'{option_name}:')
    print(f'MSE: {mse}')
    print(f'MAE: {mae}')
    print(f'RMSE: {rmse}')
    print(f'Standard Deviation: {std_dev}')
    print('\n')

    # Genera il grafico
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['RR (breaths/min)'], marker='o', linestyle='-', color='b', label='RR (breaths/min)')
    plt.axhline(y=target_rr, color='r', linestyle='--', label='Target RR = 12 breaths/min')
    plt.xlabel('Index')
    plt.ylabel('RR (breaths/min)')
    plt.title(f'Respiratory Rate Analysis - {option_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Esegui l'analisi per ogni opzione
for df, option_name in zip(dfs, option_names):
    analyze_option(df, option_name)

# Combinazione dei dati per analisi complessiva
df_combined = pd.concat(dfs, ignore_index=True)
df_combined['Error'] = df_combined['RR (breaths/min)'] - target_rr

# Calcolo delle metriche statistiche complessive
mse_total = mean_squared_error(df_combined['RR (breaths/min)'], [target_rr] * len(df_combined))
mae_total = mean_absolute_error(df_combined['RR (breaths/min)'], [target_rr] * len(df_combined))
rmse_total = np.sqrt(mse_total)
std_dev_total = np.std(df_combined['RR (breaths/min)'])

# Stampa delle metriche complessive
print('Overall Analysis:')
print(f'MSE: {mse_total}')
print(f'MAE: {mae_total}')
print(f'RMSE: {rmse_total}')
print(f'Standard Deviation: {std_dev_total}')

# Genera il grafico complessivo
plt.figure(figsize=(12, 6))
plt.plot(df_combined.index, df_combined['RR (breaths/min)'], marker='o', linestyle='-', color='b', label='RR (breaths/min)')
plt.axhline(y=target_rr, color='r', linestyle='--', label='Target RR = 12 breaths/min')
plt.xlabel('Index')
plt.ylabel('RR (breaths/min)')
plt.title('Respiratory Rate Analysis - Overall')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Funzione per ordinare correttamente i cicli
def sort_cycles(df):
    # Estrarre il numero del ciclo
    df['Cycle_Number'] = df['Cycle'].apply(lambda x: int(x.split(' ')[0][:-2]))
    
    # Ordinare per numero di ciclo
    df_sorted = df.sort_values(by=['Cycle_Number']).reset_index(drop=True)
    
    # Rimuovere la colonna ausiliaria
    df_sorted = df_sorted.drop(columns=['Cycle_Number'])
    
    return df_sorted
```

```python
df_p3_o1_f_sorted = sort_cycles(df_p3_o1)
df_p3_o2_f_sorted = sort_cycles(df_p3_o2)
df_p3_o3_f_sorted = sort_cycles(df_p3_o3)
df_p3_o4_f_sorted = sort_cycles(df_p3_o4)

df_all = pd.concat([df_p3_o1_f_sorted, df_p3_o2_f_sorted, df_p3_o3_f_sorted, df_p3_o4_f_sorted])
df_all_sorted = sort_cycles(df_all)
```

# Protocol 3 statistical analysis

```python
# Definisci il target RR per i due segmenti
target_rr_1 = 12
target_rr_2 = 15

# Lista dei dataframe ordinati delle varie opzioni
dfs_sorted = [df_p3_o1_f_sorted, df_p3_o2_f_sorted, df_p3_o3_f_sorted, df_p3_o4_f_sorted]
option_names = ['Option 1', 'Option 2', 'Option 3', 'Option 4']

# Funzione per eseguire l'analisi statistica e generare il grafico per ogni opzione
def analyze_option(df, option_name):
    df['Target RR'] = target_rr_1
    df.loc[15:, 'Target RR'] = target_rr_2
    df['Error'] = df['RR (breaths/min)'] - df['Target RR']

    # Calcolo delle metriche statistiche per i due segmenti
    mse_1 = mean_squared_error(df['RR (breaths/min)'][:5], [target_rr_1] * 5)
    mae_1 = mean_absolute_error(df['RR (breaths/min)'][:5], [target_rr_1] * 5)
    rmse_1 = np.sqrt(mse_1)
    std_dev_1 = np.std(df['RR (breaths/min)'][:5])

    mse_2 = mean_squared_error(df['RR (breaths/min)'][5:], [target_rr_2] * len(df['RR (breaths/min)'][5:]))
    mae_2 = mean_absolute_error(df['RR (breaths/min)'][5:], [target_rr_2] * len(df['RR (breaths/min)'][5:]))
    rmse_2 = np.sqrt(mse_2)
    std_dev_2 = np.std(df['RR (breaths/min)'][5:])

    # Stampa delle metriche
    print(f'{option_name} (Target 12):')
    print(f'MSE: {mse_1}')
    print(f'MAE: {mae_1}')
    print(f'RMSE: {rmse_1}')
    print(f'Standard Deviation: {std_dev_1}')
    print('\n')
    
    print(f'{option_name} (Target 15):')
    print(f'MSE: {mse_2}')
    print(f'MAE: {mae_2}')
    print(f'RMSE: {rmse_2}')
    print(f'Standard Deviation: {std_dev_2}')
    print('\n')

    # Genera il grafico
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['RR (breaths/min)'], marker='o', linestyle='-', color='b', label='RR (breaths/min)')
    plt.axhline(y=target_rr_1, xmin=0, xmax=5/len(df), color='r', linestyle='--', label='Target RR = 12 breaths/min')
    plt.axhline(y=target_rr_2, xmin=5/len(df), xmax=1, color='g', linestyle='--', label='Target RR = 15 breaths/min')
    plt.xlabel('Index')
    plt.ylabel('RR (breaths/min)')
    plt.title(f'Respiratory Rate Analysis - {option_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Esegui l'analisi per ogni opzione
for df, option_name in zip(dfs_sorted, option_names):
    analyze_option(df, option_name)

# Combinazione dei dati per analisi complessiva
df_combined = pd.concat(dfs_sorted, ignore_index=True)
df_all_sorted['Target RR'] = target_rr_1
df_all_sorted.loc[15:, 'Target RR'] = target_rr_2
df_all_sorted['Error'] = df_all_sorted['RR (breaths/min)'] - df_all_sorted['Target RR']

# Calcolo delle metriche statistiche complessive per i due segmenti
mse_total_1 = mean_squared_error(df_all_sorted['RR (breaths/min)'][:20], [target_rr_1] * 20)
mae_total_1 = mean_absolute_error(df_all_sorted['RR (breaths/min)'][:20], [target_rr_1] * 20)
rmse_total_1 = np.sqrt(mse_total_1)
std_dev_total_1 = np.std(df_all_sorted['RR (breaths/min)'][:20])

mse_total_2 = mean_squared_error(df_all_sorted['RR (breaths/min)'][20:], [target_rr_2] * len(df_all_sorted['RR (breaths/min)'][20:]))
mae_total_2 = mean_absolute_error(df_all_sorted['RR (breaths/min)'][20:], [target_rr_2] * len(df_all_sorted['RR (breaths/min)'][20:]))
rmse_total_2 = np.sqrt(mse_total_2)
std_dev_total_2 = np.std(df_all_sorted['RR (breaths/min)'][20:])

# Stampa delle metriche complessive
print('Overall Analysis (Target 12):')
print(f'MSE: {mse_total_1}')
print(f'MAE: {mae_total_1}')
print(f'RMSE: {rmse_total_1}')
print(f'Standard Deviation: {std_dev_total_1}')
print('\n')

print('Overall Analysis (Target 15):')
print(f'MSE: {mse_total_2}')
print(f'MAE: {mae_total_2}')
print(f'RMSE: {rmse_total_2}')
print(f'Standard Deviation: {std_dev_total_2}')

# Genera il grafico complessivo
plt.figure(figsize=(12, 6))
plt.plot(df_all_sorted.index, df_all_sorted['RR (breaths/min)'], marker='o', linestyle='-', color='b', label='RR (breaths/min)')
plt.axhline(y=target_rr_1, xmin=0, xmax=20/len(df_all_sorted), color='r', linestyle='--', label='Target RR = 12 breaths/min')
plt.axhline(y=target_rr_2, xmin=20/len(df_all_sorted), xmax=1, color='g', linestyle='--', label='Target RR = 15 breaths/min')
plt.xlabel('Index')
plt.ylabel('RR (breaths/min)')
plt.title('Respiratory Rate Analysis - Overall')
plt.legend()
plt.grid(True)
plt.show()
```
